```python
# determine city codes

# City Codes were manually looked up in openweathermap.org
```

```python
# run first list of 20 city codes
url="https://api.openweathermap.org/data/2.5/group?id=5039192,4887398,5391959,4671654,5344994,4460243,4699066,4704482,4726311,5131095,5186794,4726206,5392171,5308655,5099836,5317058,5809844,4509177,4700168,4719457&appid=" + weather_key + "&units=metric"
```

```python
# Using this method to load
response = requests.get(url)
response.json()
```

```python
#
response.json().keys()

#JSON response
response.json()['list']
```

```python
#
weather = response.json()['list']
df1 = json_normalize(weather)
```

```python
# run second list of city codes
url="https://api.openweathermap.org/data/2.5/group?id=5780993,5393015,5345860,4160021,4776222,5392423,5331835,4259418,4691930,5400075,5375480,5313457,5380748,5110302,5341145,4762894,5367929,5386834,5372223,5378538&appid=3a9fa2731241e7ab03bbf05564b12ba4&units=metric"
```

```python
response = requests.get(url)
response.json()
```

```python
weather = response.json()['list']
df2= json_normalize(weather)
```

```python
# run third list of 10 city codes
url="https://api.openweathermap.org/data/2.5/group?id=5808079,4499379,4722625,5416541,4890119,5350159,5123853,5393212,4671240&appid=3a9fa2731241e7ab03bbf05564b12ba4&units=metric"
```

```python
# Concatenate DataFrames
df4 = [df1, df2, df3]
result = pd.concat(df4)
```

```python
# clean up dataframe
# drop columns
result = result.drop(['weather', 'visibility', 'dt', 'id', 'coord.lon', 'coord.lat', 'sys.country', 'sys.timezone', 'sys.sunrise', 'sys.sunset'], axis=1)
```

```python
# export to csv 

outname = 'weatherfinal.csv'

outdir = './output_data'
if not os.path.exists(outdir):
    os.mkdir(outdir)

fullname = os.path.join(outdir, outname)    

result.to_csv(fullname)
```
